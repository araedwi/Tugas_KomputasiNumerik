```python
%matplotlib inline
import matplotlib.pyplot as  plt
import numpy as np
import numpy.polynomial.polynomial as poly
```

kode ini berfungsi membuat 50 titik acak dari persamaan x dan y yang telah ditentukan kemudian diimplementasikan membentuk sebuah grafik           

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y);
```

kode ini berfungsi membuat curve fitting dari persamaan yang sebelumnya telah ditentukan. kode tersebut mengimpor LinearRegression untuk membuat model menggunakan data x dan y, kemudian membuat prediksi nilai y disimpan di yfit berdasarkan 1000 titik x baru disimpan xfit. kemudian memvisualisasikan data asli sebagai scatter plot dan garis regresi yang diprediksi menjadi plot garis

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

model.fit(x[:, np.newaxis], y)


xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x,y)
plt.plot(xfit, yfit);


```

kode ini berfungsi untuk menampilkan model slope dan model intercept

```python
print ("Model Slope :" , model.coef_[0])
print ("Model intercept :", model.intercept_)
```

kode ini menghasilkan secara acak x dan y dengan pola non linear dengan fungsi sinus kemudian melakukan curve fitting menggunakan polinomial derajat 6 untuk mendekati pola data tersebut. koefisien polinomial dihitung menggunakan poly.polyfit dan kurva polynomial digambarkan dengan poly.Polynomial dan np.linspace untuk menghasilkan titik baru x_new dan y_new kemudian divisualisasikan dalam grafik

```python
rng = np.random.RandomState(1)
x  = 10 * rng.rand(50)
y = np.sin(x) + 0.1 * rng.randn(50)
degree = 6
coefs = poly.polyfit(x, y, degree)
ffit = poly.Polynomial(coefs)
x_new = np.linspace(0,10,100)
y_new = ffit(x_new)

plt.scatter(x,y, label='Data Asli' )
plt.plot (x_new, y_new, color='black', label='Kurva Polinomial')
plt.legend()
plt.title('Curve Fitting dengan Polinomial')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
